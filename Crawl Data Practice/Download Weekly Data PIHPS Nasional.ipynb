```python
import re
import os

from selenium import webdriver
import time
```

```python
chrome_path = 'C:/Users/Redata/Downloads/read/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(chrome_path)
```

```python
driver.get("https://hargapangan.id/tabel-harga")
```

```python
prov_list = driver.find_element_by_id("filter_province_ids").text.split("\n")
```

```python
komoditi_list = driver.find_element_by_id("filter_commodity_ids").text.split("\n")
```

```python
##set tanggal dan jenis laporan
# note "01-07-2016" and week to wwek
```

```python
for i in komoditi_list:
    driver.find_element_by_xpath(f"//option[contains(text(),'{i.strip()}')]").click()
    for j in prov_list:
        driver.find_element_by_xpath(f"//option[contains(text(),'{j.strip()}')]").click()
        driver.find_elements_by_xpath("//i[contains(concat(' ', normalize-space(@class), ' '), 'fa fa-file-text')]")[0].click()
        driver.implicitly_wait(10)
        filen_name = driver.find_element_by_id("report-header").text.split("\n")
        a = filen_name[1].split(":")[-1] + "_" + filen_name[2].split(":")[-1] + "_" + filen_name[-1].split(":")[-1]
        a = i.strip() + a
        driver.find_element_by_id("btnDownload").click()
        try:
            if driver.find_element_by_id("sub-frame-error") != None:
                driver.refresh()
        except:
            print ("tidak perlu")
        driver.implicitly_wait(10)
        os.rename("C:/Users/Redata/Downloads/export.xls", f"C:/Users/Redata/Downloads/pihs/{a}.xls")
        driver.find_element_by_xpath(f"//option[contains(text(),'{j.strip()}')]").click()
        driver.implicitly_wait(10)
    driver.find_element_by_xpath(f"//option[contains(text(),'{i.strip()}')]").click()
    driver.implicitly_wait(10)
```
